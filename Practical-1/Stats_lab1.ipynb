```python
import pandas as pd
df = pd.read_csv('Diwali_Sales_Data.csv', encoding='latin1')
df.head()
```

```python
!ls
```

```python
df.info()
```

```python
df = df[0:]
print(df.isna())
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
def Male_female():
    return df.groupby('Gender').count()
```

```python
Male_female()
```

```python
import matplotlib.pyplot as plt
male_female = df['Gender'].value_counts()
plt.bar(male_female.index, male_female.values, color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
male_female = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=male_female)
```

```python
df.head()
```

```python
Age_group = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=Age_group)
```

```python
Age_group_males_females = df.groupby(['Age Group','Gender'],as_index=False).sum()
sns.barplot(x='Age Group',y='Amount',hue='Gender',data=Age_group_males_females)
```

```python
age_group_males_females = df.groupby(['Age Group', 'Gender'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
```
